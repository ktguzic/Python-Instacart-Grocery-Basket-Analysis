```python
# Question #3 - orders_products_combined import

ords_prods_comb = pd.read_csv(os.path.join(path, '02 Data', 'Prepped Data', 'orders_products_combined.csv'), index_col = False)
```

```python
# Question #5 - Suitable way to combine ords_prods_comb with products

# Import products dataset

df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepped Data', 'products_checked.csv'), index_col = False)
```

```python
# checking import
df_prods.head()
```

```python
# removing Unnamed heading
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Going to combine on the product_id column

df_merged_ords_prods_comb = ords_prods_comb.merge(df_prods, on = 'product_id')
```
