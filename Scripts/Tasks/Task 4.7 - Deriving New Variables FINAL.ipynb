```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# using function on dataframe

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# checking value count of defined dataframe

df['price_range'].value_counts(dropna = False)
```

```python
# Creating a For-Loop for ranking how busy each day is
```

```python
result = []
for value in ords_prods_merge["orders_day_of_the_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result = []
for value in ords_prods_merge["time_of_order"]:
  if value == 10:
    result.append("Most orders")
  elif value == 11:
     result.append("Most orders")   
  elif value == 14:
     result.append("Most orders")  
  elif value == 15:
     result.append("Most orders")
  elif value == 3:
    result.append("Fewest orders")
  elif value == 4:
    result.append("Fewest orders")
  elif value == 2:
    result.append("Fewest orders")
  elif value == 5:
    result.append("Fewest orders")
  else:
    result.append("Average Orders")
```
